```python
import cv2 as cv
import numpy as np
```

```python
video = cv.VideoCapture("/Users/amariofausta/Downloads/WhatsApp Video 2022-06-15 at 12.40.18 PM.mp4")
```

<h2> Memainkan Video </h2>

```python
# Create a VideoCapture object and read from input file
cap = video
   
# Check if camera opened successfully
if (cap.isOpened()== False): 
  print("Error opening video  file")
   
# Read until video is completed
while(cap.isOpened()):
      
  # Capture frame-by-frame
  ret, frame = cap.read()
  if ret == True:
      
    # Display the resulting frame
    cv.imshow('Frame', frame)
   
    # Press Q on keyboard to  exit
    if cv.waitKey(25) & 0xFF == ord('q'):
      break
   
  # Break the loop
  else: 
    break
   
# When everything done, release 
# the video capture object
cap.release()
cv.destroyAllWindows()
```

<h2> Apa itu ret? </h2>
<h3> ret merupakan status pemutaran video </h3>
<ul>

    <li> Apabila hasilnuya false maka video telah selesai </li>

<h2> Mencoba menjadi HSV </h2>

```python
# Create a VideoCapture object and read from input file
cap = video
   
# Check if camera opened successfully
if (cap.isOpened()== False): 
  print("Error opening video  file")
   
# Read until video is completed
while(cap.isOpened()):
      
  # Capture frame-by-frame
  ret, frame = cap.read()
  frame = cv.cvtColor(frame, cv.COLOR_BGR2HSV)
  if ret == True:
      
    # Display the resulting frame
    cv.imshow('Frame', frame)
   
    # Press Q on keyboard to  exit
    if cv.waitKey(25) & 0xFF == ord('q'):
      break
   
  # Break the loop
  else: 
    break
   
# When everything done, release 
# the video capture object
cap.release()
cv.destroyAllWindows()
```
